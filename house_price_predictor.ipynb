```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
df = pd.read_csv("housing.csv")
```

```python
median_tota_bedrooms = df['total_bedrooms'].median()
df['total_bedrooms']=df['total_bedrooms'].fillna(median_tota_bedrooms)
```

```python
df['ocean_proximity'].value_counts()
```

```python
unique_categories = sorted(list(set(df['ocean_proximity'])))
```

```python
print(unique_categories)
```

```python
one_hot_data = []
```

```python
for word in df['ocean_proximity']:
  current_row_encoding = []
  for category in unique_categories:
    if word == category:
      current_row_encoding.append(1)
    else:
      current_row_encoding.append(0)
  one_hot_data.append(current_row_encoding)
```

```python
for i in range(len(df['ocean_proximity'].head())):
    print(f"Original: {df['ocean_proximity'][i]:<12} -> Encoded: {one_hot_data[i]}")
```

```python
encoded_df = pd.DataFrame(one_hot_data,columns=unique_categories)
```

```python
df=pd.concat([df,encoded_df],axis=1)
```
